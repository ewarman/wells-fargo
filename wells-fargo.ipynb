```python
#read in data
import json
with open('sent-msgs.json') as data_file:    
    data = json.load(data_file)
```

```python
import re

words = set()
for msg in data:
    if msg['Sent'] is 'pos' or 'neg':
        line = re.findall(r'\w+', msg['FullText'])
        line = [x.lower() for x in line]
        words = words | set(line)
        
print(len(words))
```

```python
f = open('sent-msgs.txt', 'w')
for msg in data:
    if msg['Sent'] is 'pos' or 'neg':
        f.write(msg['FullText'].lower())
        f.write('\n')
f.close()
```

```python
from collections import Counter

with open('sent-msgs.txt') as file:
    msgs = file.read()
    
words = re.findall(r'\w+', msgs)
word_counts = Counter(words)
print(word_counts)
```

```python
#Convert all messages to lowercase
for msg in data:
    if msg['Sent'] is 'pos' or 'neg':
        msg['FullText'] = msg['FullText'].lower()
```

```python
#bank mentions

#NEED TO CLASSIFY WITH TWIT HANDLE AND HASHTAG AS WELL!!
    
BankA=[]
BankB=[]
BankC=[]
BankD=[]

BankA_words = ['BankA', '#BankA', 'twit_hndl_BankA']
BankB_words = ['BankB', '#BankB', 'twit_hndl_BankB']
BankC_words = ['BankC', '#BankC', 'twit_hndl_BankC']
BankD_words = ['BankD', '#BankD', 'twit_hndl_BankD']

for item in data:
    item['BankA'] = False
    item['BankB'] = False
    item['BankC'] = False
    item['BankD'] = False
    if any(word in item['FullText'] for word in BankA_words):
        item['BankA'] = True
    if any(word in item['FullText'] for word in BankB_words):
        item['BankB'] = True
    if any(word in item['FullText'] for word in BankC_words):
        item['BankC'] = True
    if any(word in item['FullText'] for word in BankD_words):
        item['BankD'] = True
```

```python
BankA=[]
BankB=[]
BankC=[]
BankD=[]

for item in data:
    if item['BankA'] and not item['BankB'] and not item['BankC'] and not item['BankD']:
        BankA.append(item)
    if item['BankB'] and not item['BankA'] and not item['BankC'] and not item['BankD']:
        BankB.append(item)
    if item['BankC'] and not item['BankA'] and not item['BankB'] and not item['BankD']:
        BankC.append(item)
    if item['BankD'] and not item['BankA'] and not item['BankB'] and not item['BankC']:
        BankD.append(item)
        
print(len(BankA))
print(len(BankB))
print(len(BankC))
print(len(BankD))

```
